```python
import pandas as pd
from sqlalchemy import create_engine

# Connect to the Postgres database
engine = create_engine("postgresql+psycopg2://airflow:airflow@localhost:5432/airflow")

# Load the covid_data table
df = pd.read_sql("SELECT * FROM covid_data", engine)

print(df.head())

# Example: Top 10 countries by confirmed cases
top10 = df.sort_values(by="TotalConfirmed", ascending=False).head(10)
print(top10)

# Plot
top10.plot(x="Country", y="TotalConfirmed", kind="bar", figsize=(10,5), title="Top 10 Countries by COVID Cases")
```
